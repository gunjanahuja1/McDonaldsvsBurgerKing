```python
mc['Fat (g)']=mc['Total Fat']
mc['Protein (g)']=mc['Protein']
mc['Sugars (g)']=mc['Sugars']
mc['Dietary Fiber (g)']=mc['Dietary Fiber']
```

```python
mc
```

```python
def compare_parameters(parameter):
    mcdonalds_mean = mc[parameter].mean()
    burger_king_mean = bk[parameter].mean()
    
    mcdonalds_max = mc[parameter].max()
    burger_king_max = bk[parameter].max()
    
    mcdonalds_min = mc[parameter].min()
    burger_king_min = bk[parameter].min()
    
    print(f"Parameter: {parameter}")
    print("----------------------------")
    print("McDonald's:")
    print(f"  Mean: {mcdonalds_mean:.2f}")
    print(f"  Max: {mcdonalds_max}")
    print(f"  Min: {mcdonalds_min}")
    print("Burger King:")
    print(f"  Mean: {burger_king_mean:.2f}")
    print(f"  Max: {burger_king_max}")
    print(f"  Min: {burger_king_min}")
    print("\n")

parameters_to_compare = ['Calories', 'Fat (g)', 'Protein (g)', 'Sugars (g)']

for parameter in parameters_to_compare:
    compare_parameters(parameter)
```

```python
fig, axes = plt.subplots(nrows=len(parameters_to_compare), ncols=1, figsize=(8, 10))
fig.suptitle('Comparison of Parameters between McDonald\'s and Burger King', fontsize=16)

for idx, parameter in enumerate(parameters_to_compare):
    ax = axes[idx]
    
    mcdonalds_values = mc[parameter]
    burger_king_values = bk[parameter]
    
    ax.hist(mcdonalds_values, alpha=0.5, label="McDonald's", color='blue', bins=20)
    ax.hist(burger_king_values, alpha=0.5, label="Burger King", color='orange', bins=20)
    
    ax.set_title(parameter.capitalize())
    ax.set_xlabel(parameter.capitalize())
    ax.set_ylabel('Frequency')
    ax.legend()

plt.tight_layout()
plt.show()
```

# Comparing Dietary fibres

```python
plt.figure(figsize=(8, 6))
plt.hist(mc['Dietary Fiber (g)'], alpha=0.5, label="McDonald's", color='blue', bins=20)
plt.hist(bk['Dietary Fiber (g)'], alpha=0.5, label="Burger King", color='orange', bins=20)

plt.title(f'Comparison of Dietary Fiber (g) between McDonald\'s and Burger King')
plt.xlabel('Dietary Fiber (g)')
plt.ylabel('Frequency')
plt.legend()

# Show the plot
plt.show()
```
